```python
# 기존 HTML 파일 불러오기
file_path = "./blog.html"
with open(file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# 테이블을 찾고 해당 섹션에 새로운 데이터 추가
def insert_data_into_table(html, data, table_name):
    table_start = html.find(f'<h3 style="color: #000000; text-align: center;" data-ke-size="size23"><b>{table_name}</b></h3>')
    if table_start == -1:
        return html  # 해당 섹션이 없으면 그대로 반환

    table_start = html.find("<table", table_start)
    table_end = html.find("</table>", table_start) + len("</table>")

    existing_table = html[table_start:table_end]

    new_rows = ""
    for name, values in data.items():
        print(name)
        new_row = f'<tr><td>{name}</td>'
        for key in ['중량 (g)', '열량 (kcal)', '단백질 (g)', '포화지방 (g)', '당류 (g)', '나트륨 (mg)']:
            new_row += f'<td>{values.get(key, "N/A")}</td>'
        new_row += '</tr>\n'
        new_rows += new_row

    updated_table = existing_table.replace("</table>", new_rows + "</table>")
    return html.replace(existing_table, updated_table)

# 기존 테이블에 데이터 삽입
html_content = insert_data_into_table(html_content, sandwich_nutrition, "샌드위치")
html_content = insert_data_into_table(html_content, sub_salad_nutrition, "샐러드")
html_content = insert_data_into_table(html_content, sub_topping_nutrition, "토핑")
html_content = insert_data_into_table(html_content, bread_nutrition, "빵")
html_content = insert_data_into_table(html_content, cheese_nutrition, "치즈")
html_content = insert_data_into_table(html_content, sub_sauce_nutrition, "소스")
html_content = insert_data_into_table(html_content, sides_nutrition, "사이드")

# 업데이트된 HTML 파일 저장
updated_file_path = "./updated_subway_nutrition.html"
with open(updated_file_path, "w", encoding="utf-8") as file:
    file.write(html_content)

```
