```python
# ê¸°ì¡´ HTML íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
file_path = "./blog.html"
with open(file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# í…Œì´ë¸”ì„ ì°¾ê³  í•´ë‹¹ ì„¹ì…˜ì— ìƒˆë¡œìš´ ë°ì´í„° ì¶”ê°€
def insert_data_into_table(html, data, table_name):
    table_start = html.find(f'<h3 style="color: #000000; text-align: center;" data-ke-size="size23"><b>{table_name}</b></h3>')
    if table_start == -1:
        return html  # í•´ë‹¹ ì„¹ì…˜ì´ ì—†ìœ¼ë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜

    table_start = html.find("<table", table_start)
    table_end = html.find("</table>", table_start) + len("</table>")

    existing_table = html[table_start:table_end]

    new_rows = ""
    for name, values in data.items():
        print(name)
        new_row = f'<tr><td>{name}</td>'
        for key in ['ì¤‘ëŸ‰ (g)', 'ì—´ëŸ‰ (kcal)', 'ë‹¨ë°±ì§ˆ (g)', 'í¬í™”ì§€ë°© (g)', 'ë‹¹ë¥˜ (g)', 'ë‚˜íŠ¸ë¥¨ (mg)']:
            new_row += f'<td>{values.get(key, "N/A")}</td>'
        new_row += '</tr>\n'
        new_rows += new_row

    updated_table = existing_table.replace("</table>", new_rows + "</table>")
    return html.replace(existing_table, updated_table)

# ê¸°ì¡´ í…Œì´ë¸”ì— ë°ì´í„° ì‚½ì…
html_content = insert_data_into_table(html_content, sandwich_nutrition, "ìƒŒë“œìœ„ì¹˜")
html_content = insert_data_into_table(html_content, sub_salad_nutrition, "ìƒëŸ¬ë“œ")
html_content = insert_data_into_table(html_content, sub_topping_nutrition, "í† í•‘")
html_content = insert_data_into_table(html_content, bread_nutrition, "ë¹µ")
html_content = insert_data_into_table(html_content, cheese_nutrition, "ì¹˜ì¦ˆ")
html_content = insert_data_into_table(html_content, sub_sauce_nutrition, "ì†ŒìŠ¤")
html_content = insert_data_into_table(html_content, sides_nutrition, "ì‚¬ì´ë“œ")

# ì—…ë°ì´íŠ¸ëœ HTML íŒŒì¼ ì €ì¥
updated_file_path = "./updated_subway_nutrition.html"
with open(updated_file_path, "w", encoding="utf-8") as file:
    file.write(html_content)

```
